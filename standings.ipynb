```python
import requests
import lxml.html as lh
import pandas as pd
from bs4 import BeautifulSoup
```

```python
data = []
count = 1
for year in range(2019,2020):
    website_url = requests.get('https://www.basketball-reference.com/leagues/NBA_'+str(year)+'_standings.html')
    soup = BeautifulSoup(website_url.content,'lxml')
    soup.prettify()
    
    My_table = soup.find('table',{'class':"sortable stats_table"})
    tabledata = My_table.findAll('td')
    for cell in tabledata:
        data.append(cell.get_text())
    print(data)
```

```python
print(My_table)
```

```python
print(soup.prettify())
```

```python
import numpy as np
data2 = np.array(data)
refined = np.reshape(data2, (-1, 8))
```

```python
tableheader = My_table.findAll('th')
headers = []
count = 0
for item in tableheader:
    if (count < 8):
        headers.append(item.get_text())
        count=count+1
headers[0] = ''
headers
```

```python
import pandas as pd
df1 = pd.DataFrame(refined, columns=headers)
df1.head()
```

```python
col = df1.columns
cols = []
for co in col:
    cols.append(co)
cols = cols[3:]
print(cols)
```

```python
for col in cols:
    df1[col] = pd.to_numeric(df1[col], errors='coerce')
```
