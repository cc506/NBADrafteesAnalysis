```python
import requests
import lxml.html as lh
import pandas as pd
from bs4 import BeautifulSoup, Comment
```

```python
data = []
count = 1
for year in range(2000,2020):
    website_url = requests.get('http://www.draftexpress.com/league/NBA/teams/'+str(year)+'/')
    soup = BeautifulSoup(website_url.content,'lxml')
    soup.prettify()
    [tag.decompose() for tag in soup("img")]
    
    My_table = soup.find('table',{'class': "sorttable high-lows"})
    #print(My_table)
    tabledata = My_table.findAll(['td'])
    for cell in tabledata:
        count = count + 1
        data.append(cell.get_text())
    print(data)
```

```python
import numpy as np
data2 = np.array(data)
refined = np.reshape(data2, (-1, 10))
```
