```python
import requests
import lxml.html as lh
import pandas as pd
from bs4 import BeautifulSoup, Comment
import json
```

```python
players = ['admiral-schofield', 'bol-bol', 'brandon-clarke', 'bruno-fernando',
           'cameron-johnson', 'cameron-reddish', 'carsen-edwards', 'chuma-okeke', 'coby-white',
           'darius-garland', 'deandre-hunter', 'dylan-windler',
           'grant-williams', 'ignas-brazdeikis', 'isaiah-roby', 'ja-morant',
           'jalen-mcdaniels', 'jarrett-culver', 'jaxson-hayes', 'jaylen-hands',
           'jaylen-hoard', 'jaylen-nowell', 'jontay-porter', 'jordan-bone',
           'keldon-johnson', 'kevin-porter', 'kris-wilkes', 'louis-king', 
           'luguentz-dort', 'matisse-thybulle', 'mfiondu-kabengele', 'miye-oni',
           'nassir-little', 'nickeil-alexander-walker', 'naz-reid', 'nicolas-claxton', 
           'pj-washington', 'rj-barrett', 'romeo-langford', 'rui-hachimura', 'sagaba-konate',
           'tacko-fall', 'terance-mann', 'ty-jerome', 'tyler-cook', 'tyler-herro', 'zion-williamson']
length = len(players)
```

```python
data = []
item={}
count = 1
for name in players:
    website_url = requests.get('https://www.nbadraft.net/players/'+str(name)+'/')
    soup = BeautifulSoup(website_url.content,'lxml')
    soup.prettify()
    
    player_info = soup.findAll('div', {'id': ['nba_player_hd']})
    
    for container in player_info:
        item['name'] = container.h2.text
        #item['college'] = container.h2['school'].text
        item['college_img'] = container.img['src']
        data.append(item)
        
    player_info = soup.findAll('div', {'id': ['nba_player_stats']})
    
    for container in player_info:
        #item['college'] = container.h2['school'].text
        item['image'] = container.img['src']
        data.append(item)
    
```
