```python
import xlsxwriter

try:
    import urllib.request as urllib2
except ImportError:
    import urllib2
from bs4 import BeautifulSoup

#add player names here (lower case and with -) 
players = ['deandre-ayton','marvin-bagleyiii','jaren-jacksonjr','trae-young','mohamed-bamba','wendell-carterjr','collin-sexton','kevin-knox','mikal-bridges']

workbook = xlsxwriter.Workbook('collegeBball.xlsx')
worksheet = workbook.add_worksheet()
line = 2

for x in players:
    quote_page = 'https://www.sports-reference.com/cbb/players/'+ x +'-1.html'
    page = urllib2.urlopen(quote_page)
    soup = BeautifulSoup(page, 'html.parser')
    worksheet.write('A'+str(line),x)
    school_box = soup.find('td', attrs={'data-stat': 'school_name'})
    school = school_box.text.strip()
    worksheet.write('B'+str(line),school)
    games_box = soup.find('td', attrs={'data-stat': 'g'})
    games = games_box.text.strip()
    worksheet.write('C'+str(line),games)
    gs_box = soup.find('td', attrs={'data-stat': 'gs'})
    gs = gs_box.text.strip()
    worksheet.write('D'+str(line),gs)
    gs_box = soup.find('td', attrs={'data-stat': 'mp_per_g'})
    gs = gs_box.text.strip()
    worksheet.write('E'+str(line),gs)
    gs_box = soup.find('td', attrs={'data-stat': 'fg_per_g'})
    gs = gs_box.text.strip()
    worksheet.write('F'+str(line),gs)
    gs_box = soup.find('td', attrs={'data-stat': 'fga_per_g'})
    gs = gs_box.text.strip()
    worksheet.write('G'+str(line),gs)
    gs_box = soup.find('td', attrs={'data-stat': 'fg_pct'})
    gs = gs_box.text.strip()
    worksheet.write('H'+str(line),gs)
    gs_box = soup.find('td', attrs={'data-stat': 'fg2_per_g'})
    gs = gs_box.text.strip()
    worksheet.write('I'+str(line),gs)
    gs_box = soup.find('td', attrs={'data-stat': 'fg2a_per_g'})
    gs = gs_box.text.strip()
    worksheet.write('J'+str(line),gs)
    gs_box = soup.find('td', attrs={'data-stat': 'fg2_pct'})
    gs = gs_box.text.strip()
    worksheet.write('K'+str(line),gs)
    gs_box = soup.find('td', attrs={'data-stat': 'fg3_per_g'})
    gs = gs_box.text.strip()
    worksheet.write('L'+str(line),gs)
    gs_box = soup.find('td', attrs={'data-stat': 'fg3a_per_g'})
    gs = gs_box.text.strip()
    worksheet.write('M'+str(line),gs)
    gs_box = soup.find('td', attrs={'data-stat': 'fg3_pct'})
    gs = gs_box.text.strip()
    worksheet.write('N'+str(line),gs)
    gs_box = soup.find('td', attrs={'data-stat': 'ft_per_g'})
    gs = gs_box.text.strip()
    worksheet.write('O'+str(line),gs)
    gs_box = soup.find('td', attrs={'data-stat': 'fta_per_g'})
    gs = gs_box.text.strip()
    worksheet.write('P'+str(line),gs)
    gs_box = soup.find('td', attrs={'data-stat': 'ft_pct'})
    gs = gs_box.text.strip()
    worksheet.write('Q'+str(line),gs)
    gs_box = soup.find('td', attrs={'data-stat': 'orb_per_g'})
    gs = gs_box.text.strip()
    worksheet.write('R'+str(line),gs)
    gs_box = soup.find('td', attrs={'data-stat': 'drb_per_g'})
    gs = gs_box.text.strip()
    worksheet.write('S'+str(line),gs)
    gs_box = soup.find('td', attrs={'data-stat': 'trb_per_g'})
    gs = gs_box.text.strip()
    worksheet.write('T'+str(line),gs)
    gs_box = soup.find('td', attrs={'data-stat': 'ast_per_g'})
    gs = gs_box.text.strip()
    worksheet.write('U'+str(line),gs)
    gs_box = soup.find('td', attrs={'data-stat': 'stl_per_g'})
    gs = gs_box.text.strip()
    worksheet.write('V'+str(line),gs)
    gs_box = soup.find('td', attrs={'data-stat': 'blk_per_g'})
    gs = gs_box.text.strip()
    worksheet.write('W'+str(line),gs)
    gs_box = soup.find('td', attrs={'data-stat': 'tov_per_g'})
    gs = gs_box.text.strip()
    worksheet.write('X'+str(line),gs)
    gs_box = soup.find('td', attrs={'data-stat': 'pf_per_g'})
    gs = gs_box.text.strip()
    worksheet.write('Y'+str(line),gs)
    gs_box = soup.find('td', attrs={'data-stat': 'pts_per_g'})
    gs = gs_box.text.strip()
    worksheet.write('Z'+str(line),gs)
    gs_box = soup.find('td', attrs={'data-stat': 'sos'})
    gs = gs_box.text.strip()
    worksheet.write('Z'+str(line),gs)
    line+=1


workbook.close()
```
